# Band Structure Visualization for 2D Materials

This notebook demonstrates band structure calculations for:
1. **Twisted Bilayer Graphene (TBG)**
2. **Transition Metal Dichalcogenides (TMD)**

Both models use the continuum model.

```python
# Import required libraries and our custom modules
import matplotlib.pyplot as plt
import numpy as np

# Import our custom classes and functions
from TBG_bands import (
    TwistedBilayerGraphene, 
    plot_single_bandstructure as plot_single_tbg,
    plot_multiple_bandstructures as plot_multiple_tbg
)

from TMD_bands import (
    TransitionMetalDichalcogenides,
    plot_single_bandstructure as plot_single_tmd,
    plot_multiple_bandstructures as plot_multiple_tmd
)

# Set matplotlib parameters for better plots
plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.size'] = 12
```

## 1. Twisted Bilayer Graphene (TBG)

Twisted bilayer graphene exhibits fascinating physics, particularly at "magic angles" where the bands become nearly flat, leading to strong correlation effects and superconductivity.

### 1.1 Single Angle Analysis - Magic Angle

Let's start by examining the band structure at the first magic angle (θ ≈ 1.086°):

```python
# Plot band structure at the magic angle
angle = 1.086
print(f"Plotting TBG band structure at angle: {angle}°")
plot_single_tbg(angle)
```

### 1.2 Multiple Angle Comparison

```python
# Compare multiple twist angles for TBG
tbg_angles = [1.086, 0.286, 0.170]
print(f"Comparing TBG band structures at angles: {tbg_angles}°")
plot_multiple_tbg(tbg_angles)
```

## 2. Transition Metal Dichalcogenides (TMD)

TMD heterostructures show rich moiré physics with different behavior compared to graphene due to their semiconducting nature and different band structure.

### 2.1 Single Angle Analysis

```python
# Plot single TMD band structure
tmd_angle = 1.43
print(f"Plotting TMD band structure at angle: {tmd_angle}°")
plot_single_tmd(tmd_angle)
```

### 2.2 Multiple Angle Comparison

```python
# Compare multiple twist angles for TMD
tmd_angles = [1.0, 1.43, 1.67, 2.5]
print(f"Comparing TMD band structures at angles: {tmd_angles}°")
plot_multiple_tmd(tmd_angles)
```
